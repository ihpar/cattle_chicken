```python
import torch

from dataset import DataOps
from trainer_clf import TrainerClf
from clf_model import SensorClassifier
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
dops = DataOps()
```

```python
for i in range(4):
    sl, sr = i*2, i*2+1
    dl_train, dl_val, dl_test, _ = dops.build_dataloaders(
        sl, sr, batch_size=16, type="clf"
    )

    model = SensorClassifier(
        input_dim=20,
        num_classes=6
    )

    trainer = TrainerClf(
        model,
        dl_train,
        dl_val,
        dl_test,
        lr=0.001,
        epochs=200,
        patience=5,
        device=device,
        save_path=f"best_clf_model_{sl}_{sr}.pt"
    )

    print(f"\nTraining on sensors {sl} and {sr}\n")
    trainer.train()
    trainer.test()
```
